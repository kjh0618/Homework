```python
fruits = {'사과', '배', '오렌지', '수박'}
fruits.remove('사과')
print(fruits)

fruits.remove('딸기')

fruits = {'사과', '배', '오렌지', '수박'}
fruits.discard('사과')
print(fruits)
fruits.discard('딸기')
print(fruits)
```
